```python
import pandas as pd
pd.set_option('display.max_rows', None)  # Set max_rows to None to display all rows
```

```python
from finvizfinance.quote import finvizfinance

stock = finvizfinance('asan')
```

```python
stock.ticker_charts()
```

```python
stock_fundament = stock.ticker_fundament()

import pandas as pd
pd.set_option('display.max_rows', None)  # Set max_rows to None to display all rows
df_stock_fundament = pd.DataFrame.from_dict(stock_fundament, orient='index', columns=['Value'])
display(df_stock_fundament)
```

```python
outer_ratings_df = stock.ticker_outer_ratings()
print(outer_ratings_df)
```

```python
news_df = stock.ticker_news()
print(news_df)
```

```python
inside_trader_df = stock.ticker_inside_trader()
print(inside_trader_df)
```

```python
from finvizfinance.insider import Insider

finsider = Insider(option='top owner trade')
# option: latest, top week, top owner trade
# default: latest

insider_trader = finsider.get_insider()
print(insider_trader)
```
